```python
import pandas as pd
import numpy as np
import chardet

with open("/content/GYS.csv", "rb") as f:
    result = chardet.detect(f.read())
```

```python
df = pd.read_csv("/content/GYS.csv", encoding=result['encoding'])
```

```python
df.head()
```

```python
df.info()
```

```python
#cek nilai nan
print(df.isna().sum())
```

```python
# Daftar nama kolom yang ingin dihapus
columns_to_delete = ["created_year","created_month","created_date","Abbreviation","Unemployment rate","Urban_population","Latitude","Longitude"]

# Hapus kolom yang ingin dihapus
df_new = df.drop(columns=columns_to_delete)
print(df_new.isna().sum())
df_new.info()
```

```python
df_new['category'] = df_new['category'].fillna('Other')
```

```python
# merubah nilai nan pada colom channel menjadi nilai yang sering muncul
most_frequent_value = df_new["channel_type"].mode()[0]
df_new["channel_type"].fillna(most_frequent_value, inplace=True)
```

```python
# melihat update dari data
df_new
```

```python
print(df_new.isna().sum())
```

```python
# Daftar nama kolom yang ingin dihapus
columns_to_delete = ["country_rank"]

# Hapus kolom yang ingin dihapus
df_new = df_new.drop(columns=columns_to_delete)
print(df_new.isna().sum())
```

```python
df_new['Country'] = df_new['Country'].fillna('Other')
print(df_new.isna().sum())
#perubahan nilai nan pada colom contry menjadi other agar data tidak hilang tapi menjadi nilai baru
```

```python
#df_new["channel_type_rank"].fillna(df_new["channel_type_rank"].mean(), inplace=True)
df_new = df_new.dropna(axis=0, how='any', subset=['channel_type_rank'])
print(df_new.isna().sum())
#
```

```python
#df_new["video_views_for_the_last_30_days"].fillna(df_new["video_views_for_the_last_30_days"].mean(), inplace=True)
df_new = df_new.dropna(axis=0, how='any', subset=['video_views_for_the_last_30_days'])
print(df_new.isna().sum())
```

```python
df_new = df_new.dropna(axis=0, how='any', subset=['video_views_rank'])
print(df_new.isna().sum())
```
